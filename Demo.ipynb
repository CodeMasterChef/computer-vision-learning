```python
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
```

```python
pic
```

```python
pic_arr = np.asarray(pic)
pic_arr.shape
```

```python
plt.imshow(pic_arr)
```

```python
pic_red = np.copy(pic_arr)
```

```python
# RED channel. 0: pure black - 255: pure red
plt.imshow(pic_red[:, : , 0], cmap='gray')
```

```python
# Green channel
plt.imshow(pic_red[:, : , 1], cmap='gray')
```

```python
# Blue
plt.imshow(pic_red[:, : , 2], cmap='gray')
```
